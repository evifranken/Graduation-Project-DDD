```python
import pandas as pd #libarary for data manipulation and analysis
import numpy as np
```

```python
#importing dataframes
style_img_survey1 = pd.read_csv('survey1_styleLabels')
style_img_survey2 = pd.read_csv('survey2_styleLabels')
style_img_survey3 = pd.read_csv('survey3_styleLabels')
style_img_survey4 = pd.read_csv('survey4_styleLabels')

frames = [style_img_survey1, style_img_survey2, style_img_survey3, style_img_survey4]
result = pd.concat(frames)
result.reset_index(inplace=True, drop=True)

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
result
```

```python
result_s1 = pd.read_csv('result_s1')
result_s2 = pd.read_csv('result_s2')
result_s3 = pd.read_csv('result_s3')
result_s4 = pd.read_csv('result_s4')
result_images = pd.concat([result_s1, result_s2, result_s3, result_s4], axis=0)
result_images.reset_index(inplace=True, drop=True)
result_images = result_images.replace(np.nan, "Not rated")
result_images
```

```python
result_images.to_excel('result_images.xlsx', index=False)
```

```python
respondent_info_s1 = pd.read_csv('respondent_info_s1')
respondent_info_s2 = pd.read_csv('respondent_info_s2')
respondent_info_s3 = pd.read_csv('respondent_info_s3')
respondent_info_s4 = pd.read_csv('respondent_info_s4')

result_Rinfo = pd.concat([respondent_info_s1, respondent_info_s2, respondent_info_s3, respondent_info_s4], axis=0)
result_Rinfo.reset_index(inplace=True, drop=True)

result_Rinfo['Respondent ID'] = result_Rinfo['Respondent ID'].astype(int)
result_Rinfo.to_csv('result_Rinfo', index=False)

result_Rinfo
```
