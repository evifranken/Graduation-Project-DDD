### 1.0 - Data collection

Due to the lack of a complete and reliable list of fashion style adjectives or verbs that fits a consumers mental-model, an online content analysis has been performed. All data is collected by manually writing down every adjective or verb that addresses fashion style. 

Attempts have been made to increase the reliability of the data collection by choosing different types of online sources; fashion marketplaces and fashion blogs. Fashion marketplaces have been included to collect variables that  fit the consumers mental model while shopping for clothing online. The used marketplaces have been selected randomly on the basis of two criteria: popularity and range of clothing. 
In order to create more support, I have opted for more diversity by also using personal blogs. Personal blogs include words users would use themselves.
