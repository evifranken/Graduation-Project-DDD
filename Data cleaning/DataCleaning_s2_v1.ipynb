```python
import pandas as pd #libarary for data manipulation and analysis
import numpy as np
```

```python
#importing dataframes
df_s2 = pd.read_csv('survey_self-concept_version2.csv')

pd.set_option('display.max_columns', None)
df_s2.head(5)
```

```python
# Drop empty columns
df_s2 = df_s2.drop(['Collector ID', 'Start Date', 'End Date', 'IP Address', 'Email Address', 'First Name', 'Last Name', 'Custom Data 1'], axis=1)

df_s2.head(5)
```

```python
# Subset into section to easily clean
personal_info = df_s2[df_s2.columns[0:5]]
interests = df_s2.iloc[:, np.r_[0, 5:10]]
size = df_s2.iloc[:, np.r_[0, 11:22]]
sc_scale = df_s2.iloc[:, np.r_[0, 22:37]]
images = df_s2.iloc[:, np.r_[0, 37:762]]

# To dataframe
df_personal_info_s2 = pd.DataFrame(data=personal_info)
df_interests_s2 = pd.DataFrame(data=interests)
df_size_s2 = pd.DataFrame(data=size)
df_sc_scale_s2 = pd.DataFrame(data=sc_scale)
df_images_s2 = pd.DataFrame(data=images)

df_personal_info_s2
```

```python
df_personal_info_s2 = df_personal_info_s2.iloc[1: , :]

df_personal_info_s2 = df_personal_info_s2.rename(columns={'What is your age?': 'Age', 
                          'Do you live in a village or a city?': 'Residency type',
                          'In what village or city do you live?': 'Residency name',
                          'What is your occupation?': 'Occupation'})

df_personal_info_s2["Residency type"].replace({"Both (part-time village, part-time city)": "Village, City"}, inplace=True)

df_personal_info_s2.reset_index(drop=True, inplace=True)

df_personal_info_s2.head(30)
```

```python
respondentID = df_personal_info_s2[['Respondent ID']]
respondentID['Respondent ID'] = respondentID['Respondent ID'].astype(int)

respondentID
```
